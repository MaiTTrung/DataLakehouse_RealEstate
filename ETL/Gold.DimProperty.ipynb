```
%sql
use lakehouse
```

```
df = spark.read.format("delta").table("Silver").select("type","furnishing_status","legal_status","land_frontage","spacious_backyard","drive_way","land_type","apartment_status","apartment_type")
```

```
df = df.dropDuplicates(["type","furnishing_status","legal_status","land_frontage","spacious_backyard","drive_way","land_type","apartment_status","apartment_type"])
```

```
from pyspark.sql import functions as F
df = df.orderBy(F.desc("type"), F.asc("furnishing_status"))
```

```
from pyspark.sql.functions import monotonically_increasing_id

# Thêm cột ID tăng dần cho DataFrame
df = df.withColumn("id", monotonically_increasing_id())
```

```
from pyspark.sql.functions import monotonically_increasing_id


df = df.select("id","type","furnishing_status","legal_status","land_frontage","spacious_backyard","drive_way","land_type","apartment_status","apartment_type")


df = df.withColumnRenamed("id", "id_property")
```
